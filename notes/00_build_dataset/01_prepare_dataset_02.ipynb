## 前処理02

本ファイルは以下の処理を行うコードである。

1. `/workspace/datasets/statcast-customized/preprocess_01`データセットを読みだす。

2. モデル構築に利用するカラムだけを抽出した新しいサブセットを作成し、`/workspace/datasets/statcast-customized/preprocess_02`に保存する。

```python
import os
import pandas as pd
```

```python
load_root_path = '/workspace/datasets/statcast-customized-tmp/preprocess_01/'
save_root_path = '/workspace/datasets/statcast-customized-tmp/preprocess_02/'
```

```python
def get_parquet_file_path_list(load_root_path):
    file_path_list = []
    # Walk through the directory and its subdirectories to find all Parquet files
    for root, dirs, files in os.walk(load_root_path):
        for file in files:
            if file.endswith('.parquet'):
                file_path_list.append(os.path.join(root, file))
    # Sort the file paths to ensure consistent order
    file_path_list.sort()
    return file_path_list
```

```python
def load_parquet(path: str) -> pd.DataFrame:
    """Read a Parquet file and return a pandas DataFrame."""
    return pd.read_parquet(path)
```

```python
parquet_file_path_list = get_parquet_file_path_list(load_root_path)
for file_path in parquet_file_path_list:
    print(f'Load: {file_path}')
print(f'Number of Parquet files: {len(parquet_file_path_list)}')
```

```python
# １ファイルで実験してみる
df_one_month = load_parquet(parquet_file_path_list[0])
with pd.option_context('display.max_columns', None, 'display.max_rows', None):
    display(df_one_month.head(10))
```

```python
batting_dynamics_columns = [
    'description',
    'bb_type',
    'launch_speed',
    'launch_angle',
    'hit_distance_sc',
]
```

```python
pitching_dynamics_columns = [
    'p_throws',
    'pitch_type',
    'release_speed',
    'release_spin_rate',
    'pfx_x',
    'pfx_z',
    'plate_x',
    'plate_z',
]
```

```python
batter_id_columns = [
    'batter',
    'stand',
]
```

```python
game_context_columns = [
    'inning',
    'inning_topbot',
    'outs_when_up',
    'balls',
    'strikes',
    'on_1b',
    'on_2b',
    'on_3b',
    'bat_score',
    'fld_score',
    'pitch_number',
]
```

```python
# listを結合して、最終的なカラムリストを作成
columns_use = batting_dynamics_columns + pitching_dynamics_columns + batter_id_columns + game_context_columns
print(columns_use)
```

```python
# １ファイルで実験してみる
df_one_month_extracted = df_one_month[columns_use]
with pd.option_context('display.max_columns', None, 'display.width', None):
    display(df_one_month_extracted.head(10))
```

```python
def add_at_bat_id(df: pd.DataFrame, count: int):
    """
    dfと現在のcountを受け取り、at_bat_id列を先頭に追加したdfと更新後countを返す。
    at_bat_idは pitch_number == 1 を打席開始として連番を振る。
    """
    if df.empty:
        return df.copy(), count

    df_out = df.copy()

    # 既存のat_bat_idがあれば作り直す
    if "at_bat_id" in df_out.columns:
        df_out = df_out.drop(columns=["at_bat_id"])

    start_flag = df_out["pitch_number"].eq(1).astype("int64")

    # 先頭がpitch_number!=1の場合でもIDが0にならないように補正
    if start_flag.iloc[0] == 0:
        start_flag.iloc[0] = 1

    at_bat_id = start_flag.cumsum() + count - 1
    df_out.insert(0, "at_bat_id", at_bat_id)

    updated_count = int(at_bat_id.iloc[-1]) + 1
    return df_out, updated_count
```

```python
# １ファイルで実験してみる
df_one_month_with_id, updated_count = add_at_bat_id(df_one_month_extracted, count=0)
with pd.option_context('display.width', None, 'display.max_rows', None):
    display(df_one_month_with_id.head(10))
    display(df_one_month_with_id.tail(10))
print(f'Length of df_one_month_extracted: {len(df_one_month_extracted)}')
print(f'Updated count: {updated_count}')
```

```python
def get_and_save_extracted_parquet(parquet_file_path, save_root_path, columns_use, at_bat_id):
    df = load_parquet(parquet_file_path)
    df_extracted = df[columns_use]
    df_extracted_with_id, at_bat_id_updated = add_at_bat_id(df_extracted, count=at_bat_id)

    # Create the save path by replacing the load root with the save root
    relative_path = os.path.relpath(parquet_file_path, load_root_path)
    save_path = os.path.join(save_root_path, relative_path)

    # Ensure the directory exists
    os.makedirs(os.path.dirname(save_path), exist_ok=True)

    # Save the extracted DataFrame to a new Parquet file
    df_extracted_with_id.to_parquet(save_path, index=False)
    print(f'Saved extracted data to: {save_path}')
    return at_bat_id_updated
```

```python
at_bat_id = 0
for parquet_file_path in parquet_file_path_list:
    at_bat_id = get_and_save_extracted_parquet(parquet_file_path, save_root_path, columns_use, at_bat_id)
```
