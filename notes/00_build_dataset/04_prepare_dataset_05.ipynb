本ファイルは以下の処理を行うコードである。

```python
import os
import numpy as np
import pandas as pd
```

```python
load_root_path = '/workspace/datasets/statcast-customized-tmp/preprocess_04/'
save_root_path = '/workspace/datasets/statcast-customized-tmp/preprocess_05/'
```

```python
def get_parquet_file_path_list(load_root_path):
    file_path_list = []
    # Walk through the directory and its subdirectories to find all Parquet files
    for root, dirs, files in os.walk(load_root_path):
        for file in files:
            if file.endswith('.parquet'):
                file_path_list.append(os.path.join(root, file))
    # Sort the file paths to ensure consistent order
    file_path_list.sort()
    return file_path_list
```

```python
def load_parquet(path: str) -> pd.DataFrame:
    """Read a Parquet file and return a pandas DataFrame."""
    return pd.read_parquet(path)
```

```python
parquet_file_path_list = get_parquet_file_path_list(load_root_path)
for file_path in parquet_file_path_list:
    print(f'Load: {file_path}')
print(f'Number of Parquet files: {len(parquet_file_path_list)}')
```

```python
# データの確認
df_one_month = load_parquet(parquet_file_path_list[0])
with pd.option_context('display.max_columns', None, 'display.max_rows', None):
    display(df_one_month.head(20))
```

```python
# 全てのCSVファイルを読み込んで1つのDataFrameに結合
df_all = pd.concat(
    [load_parquet(path) for path in parquet_file_path_list],
    ignore_index=True
)
print(df_all.shape)
display(df_all.head())
```

```python
unique_at_bat_id_count = df_all["at_bat_id"].nunique()
print(f"at_bat_id のユニーク数: {unique_at_bat_id_count}")
print(f'最後の at_bat_id: {df_all["at_bat_id"].iloc[-1]}')
```

```python
from sklearn.model_selection import train_test_split

# at_bat_id でtrain/valid/testに分割して保存

# 全ユニーク at_bat_id を取得
all_at_bat_ids = df_all["at_bat_id"].unique().tolist()
print(f"Total unique at_bat_ids: {len(all_at_bat_ids)}")

# 分割割合
train_ratio = 0.8
valid_ratio = 0.1
test_ratio  = 0.1
SEED = 42

# train / (valid + test) に分割
train_ids, temp_ids = train_test_split(
    all_at_bat_ids,
    test_size=(valid_ratio + test_ratio),
    random_state=SEED
)

# valid / test に分割
valid_ids, test_ids = train_test_split(
    temp_ids,
    test_size=test_ratio / (valid_ratio + test_ratio),
    random_state=SEED
)

print(f"Train: {len(train_ids)} ({len(train_ids)/len(all_at_bat_ids)*100:.1f}%)")
print(f"Valid: {len(valid_ids)} ({len(valid_ids)/len(all_at_bat_ids)*100:.1f}%)")
print(f"Test:  {len(test_ids)} ({len(test_ids)/len(all_at_bat_ids)*100:.1f}%)")

# 保存
split_save_dir = os.path.join(save_root_path, "split")
os.makedirs(split_save_dir, exist_ok=True)

pd.Series(train_ids, name="at_bat_id").to_csv(os.path.join(split_save_dir, "train_at_bat_ids.csv"), index=False)
pd.Series(valid_ids, name="at_bat_id").to_csv(os.path.join(split_save_dir, "valid_at_bat_ids.csv"), index=False)
pd.Series(test_ids,  name="at_bat_id").to_csv(os.path.join(split_save_dir, "test_at_bat_ids.csv"),  index=False)

print(f"Saved split files to: {split_save_dir}")
```

```python
def create_class_label_mapping(df: pd.DataFrame, column: str):
    """
    指定カラムのユニーク値とカウント数を集計し、カウント降順でクラスラベルを付与する。

    Returns
    -------
    encoded_df : pd.DataFrame
        指定カラムをクラスラベルに置換したDataFrame
    stats_df : pd.DataFrame
        [class_label, <column>, count] を持つ集計テーブル
    """
    if column not in df.columns:
        raise ValueError(f"'{column}' はDataFrameに存在しません。")

    # NaNを除いて件数集計（降順）
    counts = df[column].value_counts(dropna=True)

    # ユニーク値・件数・クラスラベル対応表
    stats_df = pd.DataFrame({
        column: counts.index,
        "count": counts.values
    })
    stats_df["class_label"] = np.arange(len(stats_df), dtype=int)
    stats_df = stats_df[["class_label", column, "count"]]

    # NaNを安全にマッピング
    sentinel = object()
    src = df[column].astype("object").where(df[column].notna(), sentinel)
    key = stats_df[column].astype("object").where(stats_df[column].notna(), sentinel)

    label_map = pd.Series(stats_df["class_label"].to_numpy(), index=key)
    encoded_col = src.map(label_map).astype("Int64")

    encoded_df = df.copy()
    encoded_df[column] = encoded_col

    return encoded_df, stats_df
```

```python
# これまで作成した stats_* を一括で整理
stats_tables = {
    "swing_result": stats_swing_result.copy(),
    "bb_type": stats_bb_type.copy(),
    "p_throws": stats_p_throws.copy(),
    "stand": stats_stand.copy(),
    "pitch_type": stats_pitch_type.copy(),
    "batter": stats_batter.copy(),
}

# 共通フォーマットに統一して縦結合
stats_long_list = []
for feature_name, stats_df in stats_tables.items():
    value_col = [c for c in stats_df.columns if c not in ["class_label", "count"]][0]
    tmp = stats_df.rename(columns={value_col: "value"}).copy()
    tmp.insert(0, "feature", feature_name)
    stats_long_list.append(tmp[["feature", "class_label", "value", "count"]])

stats_all = pd.concat(stats_long_list, ignore_index=True)

print("stats_all shape:", stats_all.shape)
display(stats_all.head(20))

# 保存
stats_save_dir = os.path.join(save_root_path, "stats")
os.makedirs(stats_save_dir, exist_ok=True)

for feature_name, stats_df in stats_tables.items():
    stats_df.to_csv(os.path.join(stats_save_dir, f"stats_{feature_name}.csv"), index=False)

stats_all.to_csv(os.path.join(stats_save_dir, "stats_all.csv"), index=False)
print(f"Saved: {stats_save_dir}")
```

```python
def encode_features_with_stats_all(df: pd.DataFrame) -> pd.DataFrame:
    """
    df内の以下カラムを、stats_allの対応表を使ってクラスラベル（Int64）に置換したDataFrameを返す。
    - swing_result
    - bb_type
    - p_throws
    - stand
    - pitch_type
    - batter
    """
    target_features = ["swing_result", "bb_type", "p_throws", "stand", "pitch_type", "batter"]

    missing_cols = [c for c in target_features if c not in df.columns]
    if missing_cols:
        raise ValueError(f"dfに必要なカラムがありません: {missing_cols}")

    encoded_df = df.copy()
    sentinel = object()

    for feature in target_features:
        map_df = stats_all.loc[stats_all["feature"] == feature, ["value", "class_label"]].copy()
        if map_df.empty:
            raise ValueError(f"stats_allにfeature='{feature}'の定義がありません。")

        # 重複があれば先頭を採用
        map_df = map_df.drop_duplicates(subset=["value"], keep="first")

        key = map_df["value"].astype("object").where(map_df["value"].notna(), sentinel)
        label_map = pd.Series(map_df["class_label"].to_numpy(), index=key)

        src = encoded_df[feature].astype("object").where(encoded_df[feature].notna(), sentinel)
        encoded_df[feature] = src.map(label_map).astype("Int64")

    return encoded_df
```

```python
def count_missing_values(df: pd.DataFrame) -> pd.DataFrame:
    """
    各カラムの欠損値数（NaN / <NA>）を集計して返す。
    """
    if not isinstance(df, pd.DataFrame):
        raise TypeError("df must be a pandas DataFrame")

    missing_counts = df.isna().sum()

    result = (
        missing_counts.rename("missing_count")
        .reset_index()
        .rename(columns={"index": "column"})
    )
    return result
```

```python
# 実験
df_one_month_encoded = encode_features_with_stats_all(df_one_month)
with pd.option_context('display.max_columns', None, 'display.max_rows', None):
    display(df_one_month_encoded.head(20))
```

```python
# NaNの数をカウント
missing_counts = count_missing_values(df_one_month_encoded)
print(missing_counts)
```

```python
def get_features_and_save_parquet(parquet_file_path, save_root_path):
    df = load_parquet(parquet_file_path)
    df_with_features = encode_features_with_stats_all(df)

    # Create the save path by replacing the load root with the save root
    relative_path = os.path.relpath(parquet_file_path, load_root_path).replace('.parquet', '.parquet')
    save_path = os.path.join(save_root_path, relative_path)

    # Ensure the directory exists
    os.makedirs(os.path.dirname(save_path), exist_ok=True)

    # Save the extracted DataFrame to a new Parquet file
    df_with_features.to_parquet(save_path, index=False)
    print(f'Saved extracted data to: {save_path}')
```

```python
for parquet_file_path in parquet_file_path_list:
    get_features_and_save_parquet(parquet_file_path, save_root_path)
```
